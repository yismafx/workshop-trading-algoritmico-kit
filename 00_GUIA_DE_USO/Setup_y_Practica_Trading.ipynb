````python
# ═══════════════════════════════════════════════════════════
# 🔑 CONFIGURACIÓN BROKER ALPACA
# ═══════════════════════════════════════════════════════════
#
# ⚠️ ACCIÓN REQUERIDA: Reemplaza las llaves abajo
#
# ═══════════════════════════════════════════════════════════

# ⚠️ REEMPLAZA con tus llaves de Alpaca PAPER
API_KEY = "TU_API_KEY"  # ← Pega tu API Key aquí
SECRET_KEY = "TU_SECRET_KEY"  # ← Pega tu Secret Key aquí

# URL para paper trading (NO cambiar)
BASE_URL = "https://paper-api.alpaca.markets"

print("🔑 Configurando conexión con broker Alpaca...\n")

# Validar que cambiaste las llaves
if API_KEY == "TU_API_KEY" or SECRET_KEY == "TU_SECRET_KEY":
    print("❌ ERROR: Aún no configuraste tus llaves")
    print("\n🔧 Instrucciones:")
    print("   1. Ve a: https://app.alpaca.markets/paper/dashboard/overview")
    print("   2. Obtén tus llaves API")
    print("   3. Reemplaza arriba donde dice TU_API_KEY y TU_SECRET_KEY")
    print("   4. Ejecuta esta celda nuevamente")
else:
    # Test de conexión
    try:
        from alpaca.trading.client import TradingClient
        
        trading_client = TradingClient(API_KEY, SECRET_KEY, paper=True)
        account = trading_client.get_account()
        
        print("✅ ¡CONEXIÓN EXITOSA!\n")
        print("═" * 60)
        print("📊 TU CUENTA PAPER:")
        print("═" * 60)
        print(f"💰 Balance: ${float(account.cash):.2f}")
        print(f"📈 Poder de compra: ${float(account.buying_power):.2f}")
        print(f"📊 Valor portafolio: ${float(account.portfolio_value):.2f}")
        print(f"✅ Estado: {account.status}")
        print("═" * 60)
        
        print("\n🎉 ¡Configuración completada!")
        print("\n📍 En el workshop usarás estas credenciales")
        print("   para conectar tus estrategias de trading.")
        
    except Exception as e:
        print("❌ ERROR al conectar\n")
        print(f"Razón: {str(e)[:100]}\n")
        print("🔧 Soluciones:")
        print("   1. Verifica que copiaste las llaves COMPLETAS")
        print("   2. Usa llaves de PAPER (no live)")
        print("   3. Sin espacios extras al copiar")
````
